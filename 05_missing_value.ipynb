## practice
- 남자 승객의 결측치는 남자 승객의 나이 평균
- 여자 승객의 결측치는 여자 승객의 나이 평균
- 결측치를 모두 채우고 나서 모든 사람의 나이 평균

```python
df_copy = df.copy()
```

```python
con1 = df_copy['sex'] == 'male'
con2 = df_copy['sex'] == 'female'

m_age_mean = df_copy.loc[con1, 'age'].mean()
f_age_mean = df_copy.loc[con2, 'age'].mean()

df_copy.loc[con1, 'age'] = df_copy.loc[con1, 'age'].fillna(m_age_mean)
df_copy.loc[con2, 'age'] = df_copy.loc[con2, 'age'].fillna(f_age_mean)

print(df_copy['age'].mean())
```
