### practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

- 남성 흡연자 조건 필터릴
- tip을 가장 많이 낸 순서로 내림차순 정열
- 상위 10개 출력

- 식사인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
con1 = tips['sex'] == 'Male'
con2 = tips['smoker'] == 'Yes'
tips.loc[con1 & con2].sort_values('tip', ascending = False).head(10)
```

```python
con1 = tips['size'] >= 5
con2 = tips['time'] == 'Dinner'
con3 = tips['smoker'] == 'No'
con4 = tips['day'].isin(['Sat', 'Sun'])
tips.loc[con1 & con2 & con3 & con4, ['total_bill', 'tip']]
```
