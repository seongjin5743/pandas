```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

# 컬럼 추가

```python
df_copy = df.copy()
```

```python
df_copy['VIP'] = False

df_copy.head(2)
```

```python
df_copy['family'] = df_copy['sibsp'] + df_copy['parch']

df_copy.head(2)
```

```python
df_copy['gender'] = df_copy['sex'] + '-' + df_copy['who']

df_copy.head(2)
```

```python
df_copy['f/a'] = round(df_copy['fare'] / df_copy['age'], 2)

df_copy.head(2)
```

# 삭제
- 행 삭제
- 열 삭제

```python
df_copy.drop(range(5))
```

```python
df_copy.drop([1, 3, 5, 7, 9])
```

```python
# 열 삭제
df_copy.drop('VIP', axis = 1)
# df_copy.drop('VIP', axis = 'columns')
```

```python
df_copy.drop(['deck', 'VIP', 'f/a'], axis = 1, inplace = True)
df_copy
```

## cut()

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins)
```

```python
pd.cut(df_copy['운동량'], bins = 10)
pd.cut(df_copy['운동량'], bins = 10).value_counts()
```

```python
pd.qcut(df_copy['운동량'], q = 10)
pd.qcut(df_copy['운동량'], q = 10).value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels = ['적음', '보통', '많음'])
df_copy.head(2)
```
