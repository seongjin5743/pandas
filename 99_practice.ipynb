```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df_copy = df.copy()
df_copy
```

```python
# 5000g 이상의 펭귄 출력
df_copy = df.copy()
con = df['body_mass_g'] >= 5000 # 5000 이상의 조건
df_copy[con]
```

```python
# Gentoo종의 flipper_length 평균
df_copy = df.copy()
con = df_copy['species'] == 'Gentoo' # Gentoo 종이란 조건
print(df_copy.loc[con, 'flipper_length_mm'].mean()) # 평균 구하기
```

```python
# Female Chinstrap종의 bill_length 높은순 정렬
df_copy = df.copy()
con = df_copy['sex'] == 'Female' # Female이란 조건
con2 = df_copy['species'] == 'Chinstrap' # Chinstrap종이란 조건
df_copy.loc[con & con2].sort_values('bill_length_mm', ascending = False) # bill_length 높은 순 정렬
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...

df_copy = df.copy()
for species in df_copy['species'].unique(): # df_copy['species'].unique()는 ['Adelie' 'Chinstrap' 'Gentoo'] 나오기 때문에 for문을 써서 종별 구분
    for sex in df_copy['sex'].unique()[:2]: # df_copy['sex'].unique()는 ['Male' 'Female' nan] 나오기 때문에 for문을 쓰지만 nan을 쓰지 않고 Female까지만 나올 수 있도록 [:2]를 써서 성별 구분
        con1 = df_copy['species'] == species
        con2 = df_copy['sex'] == sex
        print(f'{species}, {sex}의 평균 몸무게는 ', end = '')
        print(df_copy.loc[con1 & con2, 'body_mass_g'].mean()) # 평균 몸무게 출력
```

```python
# bill_depth 결측치 채우기
# => 같은종 같은섬의 평균으로
df_copy = df.copy()

for species in df_copy['species'].unique(): # df_copy['species'].unique()는 ['Adelie' 'Chinstrap' 'Gentoo'] 나오기 때문에 for문을 써서 종별 구분
    for island in df_copy['island'].unique(): # df_copy['island'].unique()는 ['Torgersen' 'Biscoe' 'Dream'] 나오기 때문에 for문을 써서 섬별 구분
        con1 = df_copy['species'] == species
        con2 = df_copy['island'] == island
        avg = df_copy.loc[con1 & con2, 'bill_depth_mm'].mean() # 같은 종, 같은 섬의 평균
        df_copy.loc[con1 & con2, 'bill_depth_mm'] = df_copy.loc[con1 & con2, 'bill_depth_mm'].fillna(avg) # 결측치 채우기
df_copy
```
