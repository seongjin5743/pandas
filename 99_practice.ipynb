```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df_copy = df.copy()
df_copy
```

```python
# 5000g 이상의 펭귄 출력
df_copy = df.copy()
con = df['body_mass_g'] >= 5000
df_copy[con]
```

```python
# Gentoo종의 flipper_length 평균
df_copy = df.copy()
con = df_copy['species'] == 'Gentoo'
print(df_copy.loc[con, 'flipper_length_mm'].mean())
```

```python
# Female Chinstrap종의 bill_length 높은순 정렬
df_copy = df.copy()
con = df_copy['species'] == 'Chinstrap'
df_copy.loc[con].sort_values('bill_length_mm', ascending = False)
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...
df_copy = df.copy()
for species in df_copy['species'].unique():
    for sex in df_copy['sex'].unique()[:2]:
        con1 = df_copy['species'] == species
        con2 = df_copy['sex'] == sex
        print(f'{species}, {sex}의 평균 몸무게는 ', end = '')
        print(df_copy.loc[con1 & con2, 'body_mass_g'].mean())
```

```python
# bill_depth 결측치 채우기
# => 같은종 같은섬의 평균으로
df_copy = df.copy()

for species in df_copy['species'].unique():
    for island in df_copy['island'].unique():
        con1 = df_copy['species'] == species
        con2 = df_copy['island'] == island
        avg = df_copy.loc[con1 & con2, 'bill_depth_mm'].mean()
        df_copy.loc[con1 & con2, 'bill_depth_mm'] = df_copy.loc[con1 & con2, 'bill_depth_mm'].fillna(avg)
```
